# Builder Confidence

```python
# Windows absolute path
file_path = r'F:\Janaki Data science folder\janu 2023\HomeLLC2023\builderconfidence.txt'

with open(file_path, 'r', encoding='utf-8') as file:
    data = json.load(file)
```

```python
#find where in the json object is our required data
data['chartSeries'][0]['data'][0]
```

```python
# Access the "data" list and create a DataFrame
df = pd.DataFrame(data['chartSeries'][0]['data'])
df.head()
```

```python
#selecting data from 1990 since the date is in end of month format upto 2013.
df=df.loc[lambda x:x['d']>='1989-12-31T05:00:00Z',:]
df.head()
```

```python
#select columns and reset index
df=df.loc[:, lambda df: ['d', 'v']].reset_index(drop=True)

#change datatype of "d"
df['d']=pd.to_datetime(df['d'])
```

```python
df.shape
```

```python
# Add one day to each date if it's not the 1st day of the month
df['d'] = df.apply(lambda row: row['d'] + pd.DateOffset(days=1) if row['d'].day != 1 else row['d'], axis=1)
df
```

```python
# Find and display duplicate rows based on the date column
duplicate_rows = df[df.duplicated(subset=['d'], keep=False)]
duplicate_rows
```

```python
# Identify and drop the first occurrence of each duplicate based on 'date_col'
df = df[~df.duplicated(subset=['d'], keep='last')]
df[286:289]
```

```python
#reset index
df = df.reset_index(drop=True)
df[286:289]
```

```python

#select only date from column "d"
df["date"] = df["d"].dt.date

#change datatype of column 'date'
df['date']=pd.to_datetime(df['date'])

#select only required columns
BuilderConfidence_df=df.loc[:, lambda df: ['date', 'v']]

# Directly assign a new name to the column
BuilderConfidence_df.columns = ["BuilderConfidence" if col == "v" else col for col in BuilderConfidence_df.columns]
BuilderConfidence_df.dtypes
```

```python
BuilderConfidence_df
```

```python
# check for missing dates
# Generate the list of dates from 1990-01-01 to 2023-12-01 as you mentioned earlier
start_date = '1990-01-01'
end_date = '2022-12-01'
date_range = pd.date_range(start=start_date, end=end_date, freq='MS')
date_list = date_range.strftime('%Y-%m-%d').tolist()

# Convert the list of dates to a set for faster comparison
date_set = set(date_list)

# Convert the "date" column of the DataFrame to a set
date_column_set = set(BuilderConfidence_df['date'].dt.strftime('%Y-%m-%d'))

# Find the missing dates by taking the set difference
missing_dates = date_set - date_column_set

# Convert the missing dates back to a list
missing_dates_list = list(missing_dates)

print("Missing Dates:")
print(missing_dates_list)
print("date_list_len", len(date_list))
print("date_set_len", len(date_set))
```

```python
###helper function to find data in those missing dates rows
inp_date=input()
# Define the specific date you want to search for
search_date = pd.to_datetime(inp_date)

# Filter the DataFrame to find rows with the specific date
result = BuilderConfidence_df[BuilderConfidence_df['date'] == search_date]

print(result)
```
