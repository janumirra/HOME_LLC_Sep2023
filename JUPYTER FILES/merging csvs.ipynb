```python
import numpy as np
import pandas as pd
```

```python

# Define the path to the excel file
excel_file = r'F:\Janaki Data science folder\janu 2023\HomeLLC2023\Dataset\consolidated.xlsx'

# Specify the sheet names you want to read into DataFrames

sheet_names = ['HPI', 'CCI', 'CSENT','IRPI','CLI','LC_yearly','NUR','HS','DGDP','GDPGR']

# Create an empty list to store DataFrames
dfs = []

# Read each sheet into a DataFrame and append it to the list
for sheet_name in sheet_names:
    df = pd.read_excel(excel_file, sheet_name=sheet_name)
    dfs.append(df)

# Concatenate or append the DataFrames into a single DataFrame
combined_df = pd.concat(dfs, ignore_index=True)  # Use ignore_index=True to reset row indexes

# Now you have a single DataFrame 'combined_df' containing data from all sheets
combined_df
```
